```python
from qiskit import *
```

```python
from qiskit.visualization import *
```

```python
from sympy import latex
```

```python
qc = QuantumCircuit(2)
```

```python
qc.initialize([0,1],0)
```

```python
qc.initialize([1,0],1)
```

```python
qc.h(0)
```

```python
qc.draw(output='latex_source')
```

```python
backend = BasicAer.get_backend('statevector_simulator')
```

```python
result = execute(qc, backend).result()
```

```python
endvector = result.get_statevector(decimals = 3, output="latex")
```

```python
print(latex(endvector))
```

```python
plot_state_city(endvector)
```

```python
plot_bloch_multivector(endvector)
```

```python
plot_state_qsphere(endvector)
```
